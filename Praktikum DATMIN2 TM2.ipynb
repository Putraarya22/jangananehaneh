Tambahkan kode bagian (b) agar bisa menampilkan kembali kata pertama yang di lakukan one-hot encoding (hint: lakukan inverse dari hasil encode).

```python
from numpy import array
from numpy import argmax
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

docs = "I ate an apple"

# memisah kalimat menjadi token
split_docs = docs.split()
data = [doc.split() for doc in split_docs]
values = array(data).ravel()
print(values)

# integer encode
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(split_docs)
print(integer_encoded)

# binary encode
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)

# --- Inverse One-Hot ke kata pertama ---
# ambil vektor one-hot untuk kata pertama
first_word_onehot = onehot_encoded[0]
print("One-hot kata pertama:", first_word_onehot)

# cari index dari one-hot (pakai argmax)
first_word_index = argmax(first_word_onehot)

# inverse ke label aslinya
first_word = label_encoder.inverse_transform([first_word_index])
print("Kata pertama hasil inverse:", first_word[0])
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_wikipedia(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/114.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Gagal akses {url}, status code: {response.status_code}")
        return ""

    soup = BeautifulSoup(response.text, "html.parser")
    paragraphs = soup.select("div.mw-parser-output > p")

    content = []
    for p in paragraphs:
        text = p.get_text(strip=True)
        if text:
            content.append(text)

    return "\n".join(content)

# Coba scraping
url = "https://id.wikipedia.org/wiki/Saham"
saham_text = scrape_wikipedia(url)

print(saham_text)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
# --- Stopwords bahasa Indonesia ---
stop_factory = StopWordRemoverFactory()
stopwords_id = stop_factory.get_stop_words()
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
import requests
from bs4 import BeautifulSoup

def scrape_wikipedia(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/114.0.0.0 Safari/537.36"
    }

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Gagal akses {url}, status code: {response.status_code}")
        return ""

    soup = BeautifulSoup(response.text, "html.parser")
    paragraphs = soup.select("div.mw-parser-output > p")

    content = []
    for p in paragraphs:
        text = p.get_text(strip=True)
        if text:
            content.append(text)

    return "\n".join(content)

# Coba scraping
url = "https://id.wikipedia.org/wiki/Saham"
saham_text = scrape_wikipedia(url)


# --- CountVectorizer ---
count_vectorizer = CountVectorizer(stop_words=stopwords_id)
count_matrix = count_vectorizer.fit_transform([saham_text])  # pakai list

count_df = pd.DataFrame(count_matrix.toarray(),
                        columns=count_vectorizer.get_feature_names_out(),
                        index=["Saham"])

print("=== Hasil CountVectorizer ===")
print(count_df.iloc[:, :15])  # tampilkan 15 kata pertama

# --- TF-IDF ---
tfidf_vectorizer = TfidfVectorizer(stop_words=stopwords_id)
tfidf_matrix = tfidf_vectorizer.fit_transform([saham_text])  # pakai list

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(),
                        columns=tfidf_vectorizer.get_feature_names_out(),
                        index=["Saham"])

print("\n=== Hasil TF-IDF ===")
print(tfidf_df.iloc[:, :15])  # tampilkan 15 kata pertama
```

```python
import numpy as np

# Ambil daftar kata (features) dari TF-IDF
feature_names = tfidf_vectorizer.get_feature_names_out()

# Tampilkan top keywords per dokumen
def top_keywords_per_doc(X, feature_names, doc_names, top_n=10):
    for doc_idx, doc_name in enumerate(doc_names):
        print(f"\n=== Top {top_n} keywords untuk dokumen: {doc_name} ===")
        row = X[doc_idx].toarray().flatten()
        top_indices = row.argsort()[::-1][:top_n]
        for idx in top_indices:
            print(f"{feature_names[idx]} : {row[idx]:.4f}")

# Karena cuma 1 dokumen: ["Saham"]
doc_names = ["Saham"]
top_keywords_per_doc(tfidf_matrix, feature_names, doc_names, top_n=10)
```

TF-IDF (Term Frequency – Inverse Document Frequency) = skor kepentingan sebuah kata di suatu dokumen relatif terhadap semua dokumen lain.

TF (Term Frequency) → seberapa sering kata muncul di dalam 1 dokumen.

IDF (Inverse Document Frequency) → seberapa jarang kata itu muncul di semua dokumen. Semakin jarang, nilainya makin tinggi.

📌 Dokumen 1 (tentang obligasi/utang/surat) → kata utang, obligasi, surat, diterbitkan punya skor 0.4324 → artinya kata-kata itu sangat khas di dokumen ini, jarang muncul di dokumen lain.

📌 Dokumen 2 (tentang simpanan/deposito) → kata simpanan, deposito, berjangka punya skor 0.5465 → kata khas di dokumen 2.

📌 Dokumen 3 (tentang IDX/Bursa Efek Indonesia) → kata indonesia, bursa, efek, idx punya skor 0.4795 → kata khas di dokumen 3.

Kata adalah atau yang → muncul di beberapa dokumen → skornya kecil (contoh adalah cuma 0.2554, yang bahkan 0.0000). Itu karena kata umum yang tidak punya daya pembeda antar dokumen.

Modifikasi kode bagian (d) agar bisa menampilkan grafik dari tiap kata. contohnya seperti gambar di bawah.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt

text1 = ['i love nlp',
         'nlp is so cool',
         'nlp is all about helping machine language',
         'this tutorial is on basic nlp technique']

# TF-IDF
vectorizer = TfidfVectorizer()
txt_fitted = vectorizer.fit(text1)
txt_transformed = vectorizer.transform(text1)

# Ambil nilai IDF untuk setiap kata
idf = txt_fitted.idf_
words = vectorizer.get_feature_names_out()

# Plot grafik batang
plt.figure(figsize=(10,5))
plt.bar(words, idf, color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Kata")
plt.ylabel("Nilai IDF")
plt.title("Grafik Nilai IDF Tiap Kata")
plt.tight_layout()
plt.show()
```
